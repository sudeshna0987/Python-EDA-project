```python
data=dataset
plt.subplot(3, 1, 1)
sns.histplot(data['IMDb Rating'], kde=True)
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Frequency')
```

```python
data=dataset
plt.subplot(3, 1, 2)
sns.histplot(data['Duration (minutes)'], bins=15, kde=True)
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
data=dataset
top_10_directors = data['Director'].value_counts().index[:20]
plt.subplot(3, 1, 1)
sns.countplot(data=data[data['Director'].isin(top_10_directors)], x='Director', order=top_10_directors)
plt.title('Distribution of Top 10 Directors')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
dataset.head()
```

```python
data = dataset.drop(['Title','Certificates','Genre','Director','Star Cast', 'Poster-src'], axis = 1)
correlation=data.corr()
correlation['MetaScore'].sort_values(ascending=True)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=correlation, annot=True, square=True, cmap='coolwarm', linewidths=1.0)
plt.title('Correlation Matrix')
plt.show()
```

```python
data=dataset.head(20)
plt.subplot(2, 3, 1)
sns.scatterplot(data=data, x='Year', y='IMDb Rating')
plt.title('IMDb Ratings vs. Year')
plt.xlabel('Year')
plt.ylabel('IMDb Rating')

plt.subplot(2, 3, 3)
sns.scatterplot(data=data, x='MetaScore', y='IMDb Rating')
plt.title('IMDb Ratings vs. MetaScore')
plt.xlabel('MetaScore')
plt.ylabel('IMDb Rating')
```

```python
data=dataset
top_directors = data['Director'].value_counts().index[:5]
plt.subplot(3, 3, 1)
sns.boxplot(data=dataset[dataset['Director'].isin(top_directors)], x='Director', y='IMDb Rating')
plt.title('IMDb Ratings vs. Director (Top 10)')
plt.xlabel('Director')
plt.ylabel('IMDb Rating')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
data=dataset.head(50)
plt.subplot(3, 3, 2)
sns.boxplot(data=data, x='Genre', y='IMDb Rating')
plt.title('IMDb Ratings vs. Genre')
plt.xlabel('Genre')
plt.ylabel('IMDb Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data=data.head(10)
x=data['Director']
sizes=data['IMDb Rating']
plt.pie(sizes, labels=x, startangle=90, wedgeprops={"edgecolor":"black"}, autopct='%1.1f%%', explode =[0,0.1,0,0.2,0.3,0,0.5,0,0.1,0.4])
plt.show()
```
